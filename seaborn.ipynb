Ques1:Create a scatter plot to visualize the relationship between two variables, by generating a synthetic 
dataset.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic dataset
np.random.seed(42)  # For reproducibility
x = np.random.uniform(0, 100, 100)  # Random values between 0 and 100
y = 3 * x + np.random.normal(0, 30, 100)  # Linear relation with some noise

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='dodgerblue', alpha=0.7, label='Synthetic Data')
plt.title("Scatter Plot of Synthetic Dataset")
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

Ques2: Generate a dataset of random numbers. Visualize the distribution of a numerical variable.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a dataset of random numbers
data = np.random.randn(1000)  # 1000 random numbers from a normal distribution

# Visualize the distribution with a histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Distribution of Random Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Ques3: Create a dataset representing categories and their corresponding values. Compare different categories 
based on numerical values.

```python
import matplotlib.pyplot as plt

# Create dataset
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = [50, 70, 90, 60, 85]  # Numerical values for each category

# Bar chart to compare categories based on values
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color=['blue', 'orange', 'green', 'red', 'purple'], edgecolor='black')
plt.title("Comparison of Categories Based on Values")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Ques4:Generate a dataset with categories and numerical values. Visualize the distribution of a numerical 
variable across different categories.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate dataset
categories = ['Category A', 'Category B', 'Category C', 'Category D']
np.random.seed(42)  # For reproducibility

# Generate random numerical values for each category
values = {
    'Category A': np.random.normal(50, 10, 100),
    'Category B': np.random.normal(60, 15, 100),
    'Category C': np.random.normal(55, 12, 100),
    'Category D': np.random.normal(45, 20, 100),
}

# Prepare data for boxplot
data = [values[category] for category in categories]

# Boxplot to visualize the distribution of numerical values across categories
plt.figure(figsize=(8, 6))
plt.boxplot(data, tick_labels=categories, patch_artist=True, 
            boxprops=dict(facecolor='lightblue', color='black'), 
            whiskerprops=dict(color='black', linewidth=1), 
            medianprops=dict(color='red', linewidth=2))

plt.title("Distribution of Numerical Values Across Categories")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

Ques5: Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a 
heatmap.

```python
import numpy as np
import pandas as pd  # Import pandas
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a synthetic dataset with correlated features
np.random.seed(42)

# Create random data for three features
x = np.random.normal(0, 1, 1000)
y = 2 * x + np.random.normal(0, 0.5, 1000)  # y is correlated with x
z = -x + 0.5 * y + np.random.normal(0, 0.5, 1000)  # z is correlated with x and y

# Combine the features into a dataset
data = np.column_stack((x, y, z))
data = pd.DataFrame(data, columns=['Feature 1', 'Feature 2', 'Feature 3'])

# Compute the correlation matrix
corr_matrix = data.corr()

# Plot the heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, linecolor='black')
plt.title("Correlation Matrix Heatmap")
plt.show()
```
