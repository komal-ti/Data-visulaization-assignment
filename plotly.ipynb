Ques1:Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a three
dimensional space.
 np.random.seed(30)
 data = {
 'X': np.random.uniform(-10, 10, 300),

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Required for 3D plotting

# Set the seed for reproducibility
np.random.seed(30)

# Create the dataset
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}

# Create DataFrame
df = pd.DataFrame(data)

# 3D scatter plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(df['X'], df['Y'], df['Z'], color='blue', alpha=0.6)

# Labels and title
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('3D Scatter Plot of Data Points')

plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Student grades data
np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df = pd.DataFrame(data)

# Violin plot without triggering the FutureWarning
plt.figure(figsize=(8, 6))
sns.violinplot(x='Grade', y='Score', data=df, hue='Grade', palette='Set2', legend=False)
plt.title('Distribution of Scores Across Different Grade Categories')
plt.xlabel('Grade')
plt.ylabel('Score')
plt.show()
```

Ques3: Using the sales data, generate a heatmap to visualize the variation in sales across different months and 
days.
 np.random.seed(20)
 data = {
 'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
 'Day': np.random.choice(range(1, 31), 100),
 'Sales': np.random.randint(1000, 5000, 100)
 }
 
 df = pd.DataFrame(data)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sales data
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df_sales = pd.DataFrame(data)

# Pivot the data to create a matrix for the heatmap
sales_pivot = df_sales.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(sales_pivot, annot=True, cmap='YlGnBu', fmt='.0f', linewidths=0.5)
plt.title('Variation in Sales Across Different Months and Days')
plt.xlabel('Month')
plt.ylabel('Day of the Month')
plt.show()
```

Ques4: Using the given x and y data, generate a 3D surface plot to visualize the function 
x = np.linspace(-5, 5, 100)
 y = np.linspace(-5, 5, 100)
 x, y = np.meshgrid(x, y)
 z = np.sin(np.sqrt(x**2 + y**2))
 data = {
 'X': x.flatten(),
 'Y': y.flatten(),
 'Z': z.flatten()
 }
 
 df = pd.DataFrame(data)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Given x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Define the function z = sin(sqrt(x^2 + y^2))
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the data for DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}

df = pd.DataFrame(data)

# 3D Surface plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Surface plot
ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

# Labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')

plt.show()
```

Ques5:Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (x
axis), and bubble size proportional to the population.

 data = {

'Germany', 'France'],
 'Population': 
np.random.randint(100, 1000, 5),

 df = pd.DataFrame(data)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Given dataset
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Bubble chart
plt.figure(figsize=(10, 6))

# Scatter plot with bubbles
plt.scatter(df['GDP'], df['Population'], 
            s=df['Population']*10, # Bubble size proportional to population
            alpha=0.6, 
            color='blue')

# Adding labels and title
for i, country in enumerate(df['Country']):
    plt.text(df['GDP'][i], df['Population'][i], country, 
             fontsize=12, ha='center', va='center')

plt.title('Bubble Chart of Population vs GDP')
plt.xlabel('GDP (in billions)')
plt.ylabel('Population (in millions)')
plt.grid(True)

# Show plot
plt.show()
```
